```python
with zipfile.ZipFile(path, 'r') as zf:
    with zf.open('Books.csv') as f:
        df = pd.read_csv(f)

df['combined_text'] = df['Book-Title'].astype(str) + ' ' + df['Book-Author'].astype(str) + ' ' + df['Year-Of-Publication'].astype(str)

df.drop(['Book-Title', 'Book-Author', 'Year-Of-Publication','Image-URL-S','Image-URL-M','Image-URL-L'], axis=1, inplace=True)

display(df.head())
```

```python
df['combined_text']=df['combined_text'].str.lower()
```

```python
import re
```

```python
def remove_punctuations(text):
  return re.sub(r"[^\w\s]","",text)
df['combined_text'] = df['combined_text'].apply(remove_punctuations)
display(df[['combined_text']].head())
```

```python
import streamlit as st
stopwords=['is','am','are','to','a','the','this','or','not']

def remove_stopwords(combined_text):
  words=combined_text.lower().split()
  filtered=[word for word in words if word not in stopwords]
  return ' '.join(filtered)

df['combined_text']=df['combined_text'].apply(remove_stopwords)
display(df[['combined_text']])
```

```python
df['combined_text'] = df['combined_text'].fillna('')
display(df[['combined_text']].head())
```

```python
df['clean_text'] = df['combined_text']
df.drop('combined_text', axis=1, inplace=True)
display(df.head())
```
